####**Handle missing values by Imputation**

```python
# Imputation
#numerical columns (mean)
df.fillna(df.mean(), inplace=True)
#categorical columns (most frequent category)
df.fillna(df.mode().iloc[0], inplace=True)
missing_values_count = df.isnull().sum()
print(missing_values_count)
```

####**Encode categorical variables using pandas' `get_dummies()` function.**

```python
#one-hot encoding
df_encoded = pd.get_dummies(df)
print(df_encoded)
```

####**Standardize product categories if needed.**

```python
print(df['cp'].unique())
print(df['thal'].unique())
```

```python
# Standardize categorical variables - cp and thal
df['cp'] = df['cp'].map({0: 'typical angina', 1: 'atypical angina', 2: 'non-anginal pain', 3: 'asymptomatic'})
df['thal'] = df['thal'].map({0: 'normal', 1: 'fixed defect', 2: 'reversible defect'})
print("Unique values of 'cp' after standardization:")
print(df['cp'].unique())
print("\nUnique values of 'thal' after standardization:")
print(df['thal'].unique())
```

```python
#categorical columns (most frequent category)
df.fillna(df.mode().iloc[0], inplace=True)
missing_values_count = df.isnull().sum()

print(df['thal'].unique())
```

####**Visualize patient demographics using pie charts or bar plots**

```python
# Bar plot for 'exang' (Exercise-induced Angina)
plt.figure(figsize=(6, 4))
sns.countplot(x='exang', data=df)
plt.title('Exercise-induced Angina (exang)')
plt.xlabel('Exercise-induced Angina')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.show()
```

```python
# Bar plot for 'ca' (Number of Major Vessels)
plt.figure(figsize=(8, 6))
sns.countplot(x='ca', data=df)
plt.title('Number of Major Vessels (ca)')
plt.xlabel('Number of Major Vessels')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
#Bar plot for Age groups
lowest_age, highest_age = df['age'].min(), df['age'].max()
print("Lowest age:", lowest_age)
print("Highest age:", highest_age)

#age bins from 20
age_bins = [(i, i+9) for i in range(20, 80, 5)]
# Count the number of people in each age group
age_counts = [((df['age'] >= age_range[0]) & (df['age'] <= age_range[1])).sum() for age_range in age_bins]
plt.figure(figsize=(14, 6))
plt.bar(range(len(age_bins)), age_counts, tick_label=[f'{age_range[0]}-{age_range[1]}' for age_range in age_bins])
plt.title('Distribution of Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Bar plot for Serum Cholesterol (chol)
plt.figure(figsize=(10, 6))
plt.hist(df['chol'], bins=range(100, 601, 50), edgecolor='black')
plt.title('Distribution of Serum Cholesterol')
plt.xlabel('Serum Cholesterol')
plt.ylabel('Frequency')
plt.xticks(range(100, 601, 50))
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Bar plot for Resting Blood Pressure (trestbps)
plt.figure(figsize=(10, 6))
plt.hist(df['trestbps'], bins=range(90, 201, 10), edgecolor='black')
plt.title('Distribution of Resting Blood Pressure')
plt.xlabel('Resting Blood Pressure')
plt.ylabel('Frequency')
plt.xticks(range(90, 201, 10))
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
#Pie chart for Gender
gender_counts = df['sex'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=['Male', 'Female'], autopct='%1.1f%%')
plt.title('Distribution of Gender')
plt.show()
```

```python
# Pie chart for Fasting Blood Sugar (fbs)
fbs_counts = df['fbs'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(fbs_counts, labels=['Fasting > 120 mg/dl', 'Fasting <= 120 mg/dl'], autopct='%1.1f%%')
plt.title('Distribution of Fasting Blood Sugar')
plt.show()
```

```python
# Pie chart for Target (Presence of Heart Disease)
target_counts = df['target'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(target_counts, labels=['No Heart Disease', 'Heart Disease'], autopct='%1.1f%%')
plt.title('Distribution of Heart Disease')
plt.show()
```

####**Compare disease prevalence rates across different patient groups using bar plots or pie charts**

```python
def plot_disease_prevalence(data, column):
    disease_prevalence = data.groupby(column)['target'].mean() * 100
    plt.figure(figsize=(10, 6))
    disease_prevalence.plot(kind='bar')
    plt.title('Disease Prevalence Rates by {}'.format(column))
    plt.xlabel(column)
    plt.ylabel('Disease Prevalence (%)')
    plt.xticks(rotation=45)
    plt.show()
#Disease prevalence rates by gender
plot_disease_prevalence(df, 'sex')
```

```python
#Disease prevalence rates by resting electrocardiographic results
plot_disease_prevalence(df, 'restecg')
```

```python
#Disease prevalence rates by fasting blood sugar
plot_disease_prevalence(df, 'fbs')
```

```python
#Disease prevalence rates by chest pain type
plot_disease_prevalence(df, 'cp')
```
