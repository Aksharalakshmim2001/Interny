####**Remove Missing values and Outliers using pandas**

```python
#Remove missing values
df_clean = df.dropna()

#Remove outliers
def remove_outliers(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df
columns_to_check = ['Sales', 'Quantity', 'Discount', 'Profit']
df_clean = remove_outliers(df, columns_to_check)
df_clean.head()
```

####**Aggregate sales data at the desired frequency (e.g., monthly) using pandas' `resample()` function.**

```python
df_clean['Order Date'] = pd.to_datetime(df_clean['Order Date'])
df_clean.set_index('Order Date', inplace=True)
# Resample the sales data at monthly frequency
monthly_sales = df_clean['Sales'].resample('M').sum()
print(monthly_sales)
```

####**Visualize sales trends over time using Line plots.**

```python
plt.figure(figsize=(10,8))
monthly_sales.plot(kind='line', marker='o', color='red')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year'] = df['Order Date'].dt.year
yearly_sales = df.groupby('Year')['Sales'].sum()
plt.figure(figsize=(10,8))
plt.plot(yearly_sales.index, yearly_sales.values, marker='o')
plt.title('Yearly Sales')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(yearly_sales.index)
plt.show()
```

####**Compare sales performance across different product categories using bar plots.**

```python
unique_categories = df['Category'].unique()
print("Categories =", unique_categories)
```

```python
#Furniture category
furniture_df = df[df['Category'] == 'Furniture']
yearly_sales = furniture_df.groupby('Year')['Sales'].sum()
plt.figure(figsize=(10, 8))
yearly_sales.plot(kind='bar', color='green')
plt.title('Yearly Sales for Furniture Category')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Office Supplies category
office_supplies_df = df[df['Category'] == 'Office Supplies']
yearly_sales_office_supplies = office_supplies_df.groupby('Year')['Sales'].sum()
plt.figure(figsize=(10, 8))
yearly_sales_office_supplies.plot(kind='bar', color='blue')
plt.title('Yearly Sales for Office Supplies Category')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Technology category
technology_df = df[df['Category'] == 'Technology']
yearly_sales_technology = technology_df.groupby('Year')['Sales'].sum()
plt.figure(figsize=(10, 8))
yearly_sales_technology.plot(kind='bar', color='orange')
plt.title('Yearly Sales for Technology Category')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Total sales for each product category
category_sales = df.groupby('Category')['Sales'].sum()
category_sales.plot(kind='bar', figsize=(10, 6), color=['green', 'blue', 'orange'])
plt.title('Total Sales Across Different Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.grid()
plt.show()
```

###**Other Visualizations**

```python
#Total sales in each month for each year using BAR CHART
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
plt.figure(figsize=(10, 8))
monthly_sales = df.groupby(['Year', 'Month'])['Sales'].sum()
monthly_sales.unstack().plot(kind='bar', stacked=False)
plt.title('Total Sales in Each Month for Each Year')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid()
plt.show()
```

```python
#Total sales by region using PIE CHART
region_sales = df.groupby('Region')['Sales'].sum()
plt.figure(figsize=(10, 8))
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Region')
plt.show()
```

```python
# Distribution of Quantity sold using HISTOGRAM
plt.figure(figsize=(10, 8))
plt.hist(df['Quantity'], bins=7, color='pink', edgecolor='black')
plt.title('Distribution of Quantity Sold')
plt.xlabel('Quantity Sold')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
#Mode of Shipping Vs Sales using BAR CHART
plt.figure(figsize=(10, 8))
impact_analysis = df.groupby('Ship Mode')['Sales'].mean()
impact_analysis.plot(kind='bar', color='violet')
plt.title('Shipping Modes on Sales')
plt.xlabel('Shipping Mode')
plt.ylabel('Average Sales')
plt.grid()
plt.show()
```

```python
#Sales vs Profit using SCATTER PLOT
plt.figure(figsize=(10, 8))
plt.scatter(df['Sales'], df['Profit'], alpha=0.5)
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```
